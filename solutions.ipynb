```python
# Challenge 1

import string

input_text: str = "g fmnc wms bgblr rpylqjyrc gr zw fylb. rfyrq ufyr amknsrcpq ypc dmp. bmgle gr gl zw fylb gq glcddgagclr ylb rfyr'q ufw rfgq rcvr gq qm jmle. sqgle qrpgle.kyicrpylq() gq pcamkkclbcb. lmu ynnjw ml rfc spj."

rotate_by: int = 2


# Assume lower case for current purposes
# +ve rotate_by -> rotates to the right and vice-versa
def rotate(char: str, rotate_by: int) -> str:
    return chr(ord("a") + (ord(char) - ord("a") + rotate_by) % 26)


# Without usage of str.maketrans
def decipher(cipher_text: str) -> str:
    return "".join(
        rotate(char, rotate_by) if char.isalpha() else char for char in cipher_text
    )


print(decipher(input_text))


# With usage of str.maketrans
trans_table: dict[int, str] = str.maketrans(
    {char: rotate(char, rotate_by) for char in string.ascii_lowercase}
)
print(input_text.translate(trans_table))

url = "map"
print(url.translate(trans_table))
```

```python
# Challenge 5
from urllib.request import urlopen
from time import sleep
import re

nothing: int | None = 12345

max_iterations: int = 400
pattern = re.compile(r".*and the next nothing is (?P<nothing>\d+)$")


def loop(nothing: int | None):
    i: int = 0
    while i < max_iterations and nothing is not None:
        url: str = (
            f"http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing={nothing}"
        )
        with urlopen(url) as response:
            response_text: str = response.read().decode(
                response.headers.get_content_charset()
            )
            print(f"Response of url [{url}] is [{response_text}]")
            if match := pattern.match(response_text):
                nothing = int(match.group("nothing"))
            else:
                nothing = None
        sleep(1)
        i += 1


# loop(nothing)

# nothing = 16044 // 2
# loop(nothing)

nothing = 63579
loop(nothing)
```

```python
# Challenge 6
from urllib.request import urlopen
import io
import zipfile
import re

url: str = "http://www.pythonchallenge.com/pc/def/channel.zip"

pattern: re.Pattern = re.compile(r"Next nothing is (?P<nothing>\d+)")


def read_file(archive: zipfile.ZipFile, file_name: str) -> str:
    return archive.read(file_name).decode("utf-8")


with urlopen(url) as response:
    with zipfile.ZipFile(io.BytesIO(response.read())) as archive:
        # archive.printdir()

        # print(read_file(archive, 'readme.txt'))

        # From readme.txt
        nothing: int = 90052

        comments: str = ""

        while True:
            file_name: str = f"{nothing}.txt"
            file_info: zipfile.ZipInfo = archive.getinfo(file_name)
            comments += file_info.comment.decode("utf-8")
            file_content: str = read_file(archive, file_name)
            # print(f"File content of [{file_name}] is [{file_content}]")
            if match := pattern.match(file_content):
                nothing = int(match.group("nothing"))
            else:
                break

        print(comments)
        
# Hint: It's what we can breathe from the air. 
# Look at the letters might indicate that the keyword length should match with hockey string's length
```
